```python
#www.znak.ru/all/
import urllib.request
from bs4 import BeautifulSoup
import re
import requests
from datetime import date, timedelta

import psycopg2
from psycopg2 import Error

date_while = date.today()
xx = date_while.strftime('%Y-%m-%d')

channal = "https://www.znak.com/all/"
page = urllib.request.urlopen(channal)
soup = BeautifulSoup(page)

h = soup.find_all('a', {'class': 'pub'})
for h1 in h:
    if re.split('/',h1.get('href'))[1] == xx:
        post = "https://www.znak.com" + h1.get('href')
        page = urllib.request.urlopen(post)
        soup = BeautifulSoup(page)
        t = soup.find('article', {'class': ''})
        text = soup.find('section', {'class': 'header'}).text + "\n"
        t1 = t.find_all('p', {'class': ''})
        for t2 in t1:
            text = text + t2.text
        try:
            img = soup.find('div', {'class': 'firstimgcover'}).get("data-downloadurl")
        except:
            img = None
        date = soup.find('time').get('datetime')
        d = re.findall('\d+', date)
        date_time = d[0]+'-'+d[1]+'-'+d[2]+' '+d[3]+':'+d[4]

        try:
            connection = psycopg2.connect(user="tester",
                                          password="postgres",
                                          host="127.0.0.1",
                                          port="5432",
                                          database="testdb")

            cursor = connection.cursor()
            # Fetch all
            cursor.execute("SELECT * FROM sites;")
            record = cursor.fetchall()
            print("Result ", "recorded")
            # Executing a SQL query to insert data into  table
            insert_query = """INSERT INTO sites (URLHREF, DATE_TIME, TEXT, URLIMG, CHANNAL) 
            VALUES (%s, %s, %s, %s, %s)"""
            values = (post, date_time, text, img, channal)
            cursor.execute(insert_query, values)
            connection.commit()
            print("1 Record inserted successfully")

        except (Exception, psycopg2.Error) as error:
            print("Error while connecting to PostgreSQL", error)
        finally:
            if connection:
                cursor.close()
                connection.close()
                print("PostgreSQL connection is closed")
```
